```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import skimage.io
import skimage.viewer
```

```python

def img_mean(img):
    return np.mean(img)
def img_std_dev(img):
    return np.std(img)
def rgb2gray(rgb):
    '''Converts image to grayscale'''
    return np.dot(rgb[...,:3], [0.2125, 0.7154, 0.0721])
def log_img(img):
    c = 255 / np.log(1 + np.max(img)) 
    log_image = c * (np.log(img + 1)) 
    # Specify the data type so that 
    # float value will be converted to int 
    log_image = np.array(log_image, dtype = np.uint8) 
    return log_image

def gamma_correct_img(img,gamma=1):
    #convert img to grayscale
    #set default gamma to 1
    img = rgb2gray(img)
    return np.array(255*(img / 255) ** gamma, dtype = 'uint8')

def pixelVal(pix, r1, s1, r2, s2=255):
    #s2=255, max value of pixel
    if (0 <= pix and pix <= r1): 
        return (s1 / r1)*pix 
    elif (r1 < pix and pix <= r2): 
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1 
    else: 
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2 
    
def contrast_stretch_img(img,r1,s1,r2,s2=255):
    pixelVal_vec = np.vectorize(pixelVal)
    return pixelVal_vec(img, r1, s1, r2, s2)

def approx_SNR(img):
    return np.mean(img)/np.std(img)
def getRed(img):
    #returns the red aspect of the image
    red_img = img.copy()
    red_img[:, :, 1] = 0    # Zero out contribution from green & blue
    red_img[:, :, 2] = 0 
    return red_img
def getBlue(img):
    #returns the blue aspect of the image
    blue_img = img.copy() # Zero out contribution from green & red
    blue_img[:,:,0]=0
    blue_img[:,:,1]=0
    return blue_img
def getGreen(img):
    #returns the green aspect of the image
    green_img = img.copy() # Zero out contribution from red & blue
    green_img[:,:,0]=0
    green_img[:,:,2]=0
    return green_img
def colorHistogram(img):
    colors = ("r", "g", "b")
    channel_ids = (0,1,2)

    # create the histogram plot, with three lines, one for
    # each color
    fig = plt.xlim([0, 256])
    for channel_id, c in zip(channel_ids, colors):
        histogram, bin_edges = np.histogram(img[:, :, channel_id], bins=256, range=(0, 256))
        fig = plt.plot(bin_edges[0:-1], histogram, color=c)

    fig = plt.xlabel("Color value")
    fig = plt.ylabel("Pixels")

    fig = plt.show()
    return fig
def grayScaleHistogram(img):
    '''Returns a Matplotlib figure of a gray scale histogram of the image passed in'''
    
    img = rgb2gray(img)
    histogram, bin_edges = np.histogram(img, bins=256)
    fig = plt.figure()
    fig = plt.title('Grayscale Histogram')
    fig = plt.xlabel('grayscale value')
    fig = plt.ylabel('pixels')
    fig = plt.xlim(0,255)
    fig = plt.plot(bin_edges[0:-1], histogram)  
    fig = plt.show()
    return fig
    

def greenHist(img):
    #creats histogram of the green aspect of the image
    histogram, bin_edges = np.histogram(img[:, :, 1], bins=256, range=(0, 256))
    fig = plt.plot(bin_edges[0:-1], histogram, color='g')
    return fig
def blueHist(img):
    #creats histogram of the blue aspect of the image
    histogram, bin_edges = np.histogram(img[:, :, 2], bins=256, range=(0, 256))
    fig = plt.plot(bin_edges[0:-1], histogram, color='b')
    return fig
def redHist(img):
    #creats histogram of the red aspect of the image
    histogram, bin_edges = np.histogram(img[:, :, 0], bins=256, range=(0, 256))
    fig = plt.plot(bin_edges[0:-1], histogram, color='r')
    return fig

def histogramEqualization(imgName,show_histograms=True):
    #imgName = file location of image
    #show_histograms = True to show the before and after histograms
    #returns equalized image to be printed using "plt.imshow(..., cmap='gray')"
    img = cv2.imread(imgName,0)

    hist,bins = np.histogram(img.flatten(),256,[0,256])

    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max()/ cdf.max()
    if show_histograms:
        plt.title('Original Image Histogram')
        plt.plot(cdf_normalized, color = 'b')
        plt.hist(img.flatten(),256,[0,256], color = 'r')
        plt.xlim([0,256])
        plt.legend(('cdf','histogram'), loc = 'upper left')
        plt.show()
    cdf_m = np.ma.masked_equal(cdf,0)
    cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
    cdf = np.ma.filled(cdf_m,0).astype('uint8')
    equalizedImg = cdf[img]
    if show_histograms:
        hist,bins = np.histogram(img2.flatten(),256,[0,256])

        cdf = hist.cumsum()
        cdf_normalized = cdf * hist.max()/ cdf.max()

        plt.plot(cdf_normalized, color = 'b')
        plt.title('Equalized Histogram')
        plt.hist(img2.flatten(),256,[0,256], color = 'r')
        plt.xlim([0,256])
        plt.legend(('cdf','histogram'), loc = 'upper left')
        plt.show()
    return equalizedImg
    
def CLAHE(imgName):
    #returns equalized image 
    img = cv2.imread(imgName,0)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl1 = clahe.apply(img)
    plt.imshow(cl1,cmap='gray')
    return cl1
def noisy(noise_typ,image):

    if noise_typ == "gauss":
            row,col,ch= image.shape
            mean = 0
            #var = 0.1
            #sigma = var**0.5
            gauss = np.random.normal(mean,1,(row,col,ch))
            gauss = gauss.reshape(row,col,ch)
            noisy = image + gauss
            return noisy
        elif noise_typ == "s&p":
            row,col,ch = image.shape
            s_vs_p = 0.5
            amount = 0.004
            out = image
            # Salt mode
            num_salt = np.ceil(amount * image.size * s_vs_p)
            coords = [np.random.randint(0, i - 1, int(num_salt))
                      for i in image.shape]
            out[coords] = 1

            # Pepper mode
            num_pepper = np.ceil(amount* image.size * (1. - s_vs_p))
            coords = [np.random.randint(0, i - 1, int(num_pepper))
                      for i in image.shape]
            out[coords] = 0
            return out
        elif noise_typ == "poisson":
            vals = len(np.unique(image))
            vals = 2 ** np.ceil(np.log2(vals))
            noisy = np.random.poisson(image * vals) / float(vals)
            return noisy
        elif noise_typ =="speckle":
            row,col,ch = image.shape
            gauss = np.random.randn(row,col,ch)
            gauss = gauss.reshape(row,col,ch)        
            noisy = image + image * gauss
            return noisy
  
```

```python
def max_min_contrast(imgName):
    #read in as grayscale
    img = cv2.imread(imgName,0)
    minmax_img = np.zeros((img.shape[0],img.shape[1]),dtype = 'uint8')
 
    # Loop over the image and apply Min-Max formulae
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            minmax_img[i,j] = 255*((img[i,j]-np.min(img))/(np.max(img)-np.min(img)))

    return minmax_img
```
